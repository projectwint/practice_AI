```python
from PIL import Image
import os, glob, sys, numpy as np
from sklearn.model_selection import train_test_split
from keras.utils import np_utils

img_dir = '/content/drive/MyDrive/실제실행용' #이미지 경로
categories = ['water', 'ground']
np_classes = len(categories)

#이미지 크기: 128x128 px
image_w = 128
image_h = 128

pixel = image_h*image_w*3

X = []
y = []

for idx, cat in enumerate(categories):
  img_dir_detail = img_dir + "/" + cat
  files = glob.glob(img_dir_detail + '/**' +"/*.jpg", recursive=True)

  for i, f in enumerate(files):
    try:
      img = Image.open(f)
      img = img.resize((image_w, image_h))
      img = img.convert("RGB")
      data = np.asarray(img)
      #Y는 0, 1이기 때문에 idx 값으로 넣는다.
      X.append(data)
      y.append(idx)
      if i % 300 == 0:
        print(cat, " : ", f)
    except:
      print(cat, str(i) + "번째에서 에러")

X = np.array(X)
Y = np.array(y)

#print("files", files)
print("len files", len(files))
print()
print("Y", Y)
print("Y에서 0(water)의 개수 ::", len(Y[Y == 0])) #len(array[array = 값])은 해당 값이 배열에 몇 개 들어있는지를 반환해준다. 
print("Y에서 1(ground)의 개수 ::", len(Y[Y == 1]))

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1)
xy = (X_train, X_test, Y_train, Y_test)
np.save('/content/drive/MyDrive/train_set', xy)
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np

image_w = 128
image_h = 128

#이미지 불러오기
X_train, X_test, y_train, y_test = np.load('/content/drive/MyDrive/train_set.npy', allow_pickle=True)
X_train = X_train.astype(float) 
X_test = X_test.astype(float)

#가중치 계산 (두 데이터의 양이 같지 않으므로 비율을 맞춰줌)
w0 = np.mean(y_train)
w1 = 1 - w0
class_weights = {0: w0, 1: w1}

print(X_train.shape)
print(X_train.shape[0])
print(np.bincount(y_train))
print(np.bincount(y_test))
```

***import keras.backend.tensorflow_backend as K에서 ModuleNotFoundError: No module named 'keras.backend.tensorflow_backend';*** <br><br>

**keras.backend.tensorflow_backend** 라는 모듈이 찾을 수 없다면서 제대로 import 되지 않았다. <br>
추론해본 결과 tensorflow의 버전이 달라서였는데, 앞서 말한 모듈은은 tensorflow 버전 1에서 작동하는 것이었다. keras는 uninstall하고 원하는 버전을 다시 설치할 수 있었지만, tensorflow는 버전 1을 더이상 지원하지 않는다고 에러 메세지가 나왔다. <br><br>
https://patiencelee.tistory.com/m/674 에서 나온 2번째 방법대로 버전 1로 호환 + 버전2로 동작을 막는 명령어를 넣어봤지만 다시 import를 할 때는 **'keras.backend' is not a package** 라는 에러 메세지만 출력됬고, gpu 사용 설정 관련으로 import한 것이기 때문에 이번 프로젝트에서 필수적인 요소는 아니라고 판단했다. <br>
따라서 관련 코드를 삭제한 후 UI 중에 리소스 관리에서 직접 설정하기로 결정했다. 

```python
import os

#CNN에서 컨볼루션 레이어를 거치면 주요 특징만 추출돼서 학습됨. 
model = Sequential()
model.add(Conv2D(32, (3, 3), padding = "same", input_shape = (image_w, image_h, 3), activation = "relu"))
#Conv2D를 이용
#3x3 크기의 컨볼루션 레이어를 32개의 필터수만큼 처음에 생성한다.
#입력형태(input_shpae)는 128 x 128이기 때문에 (128, 128, 3)의 tuple 값을 가짐
#활성화 함수는 relu를 사용
model.add(MaxPooling2D(pool_size = (2, 2)))
#MaxPooling2D: 주요값을 뽑아내서 작은 출력값을 만들고, 사소한 변화는 무시한다. 

model.add(Conv2D(128, (3, 3), padding = "same", activation = "relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, (3, 3), padding = "same", activation = "relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(1, activation = "sigmoid"))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#2진분류(카테고리가 2개)기 때문에 binary 값으로 출력을 지정한다. 

model_dir = '/content/drive/MyDrive/model'

if not os.path.exists(model_dir):
  os.mkdir(model_dir)
model_path = model_dir + "/ground_water_classify.model"

checkpoint = ModelCheckpoint(filepath = model_path, monitor = 'val_loss', verbos = 1, save_best_only = True)
early_stopping = EarlyStopping(monitor = 'val_loss', patience = 5)
#val_loss 값이 5번 이상 상승되지 않으면 자동으로 멈추는 코드(overfitting 방지)
```

```python
history = model.fit(X_train, y_train, epochs=100, class_weight = class_weights, batch_size=128, validation_split = 0.20, callbacks = [checkpoint, early_stopping])
```
